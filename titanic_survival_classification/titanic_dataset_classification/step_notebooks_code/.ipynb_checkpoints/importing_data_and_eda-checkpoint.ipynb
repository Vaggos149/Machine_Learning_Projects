# EDA

### Survived

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

* First, let's create a barplot to see how many people died and how many survived. Will give us an intuition about the problem and will let us know about dataset imbalance

```python
df['Survived'].value_counts()
```

```python
total = float(len(df)) 
ax = sns.countplot(x="Survived", data=df) 
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}'.format(height/total),
            ha="right")
plt.title('Countplot of Survived vs No-Survived')
plt.show()
```

* So, our dataset is slightly imbalanced....We might need to fix that

### SEX

```python
df.loc[df['Age']<16,'Sex']='child'
```

* Now, we check the value counts

```python
df['Sex'].value_counts()
```

```python
ax = sns.countplot(x="Sex", data=df,hue="Survived") 
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:.0f}'.format(height),
            ha="right")
plt.title('Sex/Child Survived vs No-Survived counts')
plt.show()
```
